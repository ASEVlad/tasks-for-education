```python
import pandas as pd
import numpy as np
import seaborn as sns
from google.colab import files
import matplotlib.pyplot as plt
```

```python
upload = files.upload()
data = pd.read_csv('Admission_Predict.csv', error_bad_lines=False)
print(data.info())
```

```python
data
```

```python
sns.barplot(x='SOP', y ='GRE Score', estimator=max, hue='University Rating', order=[1.5,2.0,3.0,3.5], data=data)
```

```python
sns.distplot(data['GRE Score'])
```

```python
sns.boxplot(y = 'Chance of Admit ', x = 'University Rating', data=data)
sns.stripplot(y = 'Chance of Admit ', x = 'University Rating', data=data, size=5)
```

```python
sns.swarmplot(data['GRE Score'],data['Chance of Admit '])
```

```python
sns.stripplot(y='GRE Score', x='SOP', hue = 'University Rating', data=data, linewidth=1.2, jitter=True, split=True)
```

```python
sns.stripplot(y='GRE Score', x='SOP', data=data, size=15, hue='University Rating', jitter=True, alpha=0.7, marker='D')
```

```python
sns.violinplot(y = 'Chance of Admit ', x = 'University Rating', orient='v', data=data)
```

```python
x = sns.PairGrid(data, vars=['Chance of Admit ', 'CGPA', 'GRE Score', 'TOEFL Score'], hue='University Rating', hue_kws={'marker':['D','s','+', "o", "x"]})
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(data, x_vars=['Chance of Admit ', 'CGPA'], y_vars=['GRE Score', 'TOEFL Score'])
x = x.map(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(data=data, vars=['Chance of Admit ', 'CGPA', 'GRE Score', 'TOEFL Score'], hue='University Rating')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

```python
corr = data.iloc[:, 1:5].corr()
plt.figure(figsize=(11,11))
sns.heatmap(corr, annot=True, square=True)
```
